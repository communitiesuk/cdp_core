# **Medallion layers**
## Source data

## Bronze — Raw Ingestion Layer
The Bronze layer contains the original raw data exactly as received from source systems, without modification. It preserves the full fidelity of the incoming files or database extracts, including any duplicates, missing values, inconsistent formats, or schema drift. Bronze is primarily for traceability and auditing: it ensures the platform always retains an immutable copy of the source data and enables data engineers to reprocess or debug issues without pulling data again from the original system.

## Silver — Cleaned and Standardised Data Layer
The Silver layer holds clean, validated, and standardised versions of the Bronze data. This is where data engineers correct types, fix timestamps, handle missing values, normalise units, deduplicate records, and apply basic, domain-aware enrichments such as lookups or metadata tagging. The structure and granularity of the data remain close to the source, but the content is now reliable, queryable, and ready for analytical use. Silver tables form the trusted foundation for all downstream modelling.

## Gold — Business-Ready Analytical Layer
The Gold layer contains curated, joined, aggregated, and modelled datasets optimised for analytics, dashboards, and decision-making. Here, Silver tables are transformed into domain-specific products — such as time-series aggregates, summary metrics, dimensional models, or wide tables shaped for BI tools. Gold tables reflect defined business concepts and KPIs, enabling consistent reporting across teams. This layer is what analysts, data consumers, and SLTs rely on directly to answer questions and drive policy decisions.
